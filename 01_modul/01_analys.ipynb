```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_excel('data.xlsx', sheet_name=2)
df2 = pd.read_excel('data.xlsx', sheet_name=3)
```

```python
df['Сумма по компонентам'] = df['Сумма по компонентам'].apply(lambda x: int(x))
df['Сумма по главным компонентам, руб'] = df['Сумма по главным компонентам, руб'].apply(lambda x: int(x))
df['Статус оплаты'].fillna('нет данных', inplace=True)
df['DealID'].fillna(999, inplace=True)
```

```python
print('Сумма по главным компонентам за период с {0} по {1} составляет: {2:,} руб'.format(df['Дата'].min().strftime('%d.%m.%Y'),
                                                                               df['Дата'].max().strftime('%d.%m.%Y'),
                                                                               df['Сумма по главным компонентам, руб'].sum()))
```

<b>Сумма по главным компонентам за период с 23.08.2019 по 10.10.2019 составляет: 10,031,783 руб
